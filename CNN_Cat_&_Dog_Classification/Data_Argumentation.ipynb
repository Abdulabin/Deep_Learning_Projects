# Data Argumentation

## Data augmentation is a technique used to artificially increase the size of a training dataset by applying various transformations to the existing data. 

### Importing Dependency

```python
import os
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
```

```python
path = os.path.join(r"C:\Users\abdul\Downloads\PROJECTS\Deep_Learning_Project\Cat & dog class CNN\training_set\dogs\dog.99.jpg")
img_path = os.path.join(path)
plt.subplot(1,2,1)
img = image.load_img(img_path,target_size=(256,256),color_mode = "grayscale")
plt.imshow(img)
plt.subplot(1,2,2)
img = image.load_img(img_path,target_size=(256,256))
plt.imshow(img)
plt.show()
```

```python
img_arr = image.img_to_array(img)
print(img_arr.shape)
```

```python
final_img = img_arr.reshape(1,256,256,3)
final_img.shape
```

```python
save_path=os.path.join(r"C:\Users\abdul\Downloads\PROJECTS\Deep_Learning_Project\Cat & dog class CNN\Arg_data")
```

```python
datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')
```

```python
i = 0
for batch in datagen.flow(final_img,
                          save_to_dir=save_path, save_prefix='cat', save_format='jpeg'):
    i += 1
    if i >= 5:
        break  
```

## Making code more productive by making it a Function 

```python
def argumentation(data_path,save_path,no_of_copies,save_name):
    for x in os.listdir(data_path):
        path  = f"{data_path}\{x}"
        img_ = image.load_img(path,target_size=(256,256),)
        img_arr_ = image.img_to_array(img_)
        final_img_ = img_arr_.reshape((1,) + img_arr_.shape)

        i = 0
        for batch in datagen.flow(final_img_,
                              save_to_dir=save_path, save_prefix=save_name, save_format='jpeg'):
            i += 1
            if i >= no_of_copies:
                break  
    print(" *** Argumentation Done *** ")
```

### Dog Images Argumentation

```python
data_dog = os.path.join(r"C:\Users\abdul\Downloads\PROJECTS\Deep_Learning_Project\Cat & dog class CNN\training_set\dogs")
save_dog_path=os.path.join(r"C:\Users\abdul\Downloads\PROJECTS\Deep_Learning_Project\Cat & dog class CNN\Arg_data\Dog")
```

```python
print("Number of Image before Argumentation : ",len(os.listdir(data_dog)))
```

```python
argumentation( data_path=data_dog , save_path=save_dog_path ,  no_of_copies=5 , save_name="dog")
```

```python
print("Number of Image after Argumentation : ",len(os.listdir(save_dog_path)))
```

### Cat Images Argumentation

```python
data_cat = os.path.join(r"C:\Users\abdul\Downloads\PROJECTS\Deep_Learning_Project\Cat & dog class CNN\training_set\cats")
save_cat_path=os.path.join(r"C:\Users\abdul\Downloads\PROJECTS\Deep_Learning_Project\Cat & dog class CNN\Arg_data\Cat")
```
